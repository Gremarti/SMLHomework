# Comparison of differents Learning Methods

```python
import matplotlib.pyplot as pyplot
import seaborn
import pandas

# Insert the working directory here.
workspace = "/home/toshuumilia/Workspace/SML/"
```

## Basic Comparison: Decision Tree vs Logistic Regression vs Neural Network

*Explanation of Figure 1*

```python
# Basic Comparison Experiment #

experimentBasicMethodsDF = pandas.read_csv(workspace + "results/experimentBasicMethods.csv")

seaborn.set_style("whitegrid")
pyplot.figure(figsize=(10,5))
seaborn.barplot(x="Value", y="Measure", hue="Method",
                data=experimentBasicMethodsDF)
pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
pyplot.ylabel('Measure', fontsize=12)
pyplot.xlabel('Value', fontsize=12)
pyplot.xlim(0.5, 1)
pyplot.title('Figure 1: Comparison of learning methods with default parameters', fontsize=15)
pyplot.show()
```
